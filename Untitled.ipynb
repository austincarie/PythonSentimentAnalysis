```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
number = 0
emptyline=[]
for row in df['text']:
    clean_tweet = ""
    tweet = df['text'][number]
    tokenized_tweet = tweet.split()
    for x in tokenized_tweet:
        if(len(x) >= 20):
            tokenized_tweet.remove(x)
    for x in tokenized_tweet:
        clean_tweet += (x + " ")
    df.replace(df['text'][number], clean_tweet) # Find and Replace
    print(clean_tweet)
    vs=analyzer.polarity_scores(row)
    emptyline.append(vs)
    number += 1
df_sentiments=pd.DataFrame(emptyline)
df_sentiments.head()
```
